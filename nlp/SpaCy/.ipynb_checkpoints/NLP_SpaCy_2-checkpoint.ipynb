# Initialisation de SPACY en Français

```python
import spacy
import pandas as pd
from spacy import displacy
nlp = spacy.load('fr')
```

```python
data_all = pd.read_csv("../../datasources/jeuxvideos/meilleursjeuvideo.csv")
```

```python
data_all.head(5)
```

```python
gow = data_all['Description'][0]
```

```python
doc = nlp(gow)
for token in doc:
    print("{0}\t{1}\t{2}\t{3}\t{4}\t{5}\t{6}\t{7}\t{8}".format(
        token.text,
        token.idx,
        token.lemma_,
        token.is_punct,
        token.is_space,
        token.shape_,
        token.pos_,
        token.tag_,
        token.ent_type_
    ))
```

```python
gow.str
```

```python
for token in doc:
    print(token.text, " -> ", token.tag_)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
doc
```

```python
for chunk in doc.noun_chunks:
    print(chunk.text, " --> ", chunk.label_)
```

```python
displacy.render(doc, style='dep', jupyter=True, options={'distance': 60})
```
