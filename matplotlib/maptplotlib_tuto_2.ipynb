```python
import pandas as pd
import matplotlib.pyplot as plt
# prenons seulement les 5 premières lignes du jeu de données
titanic = pd.read_csv("../datasources/titanic/train.csv")[:5]
```

## Histogrammes

Seul un jeu de données en entrée, la librairie fait un décompte des données pour calculer la valeur (compte) en ordonnée.

```python
titanic.Fare
```

```python
plt.hist(titanic.Fare)
```

## Diagramme à barre

```python
plt.bar(titanic.PassengerId, titanic.Fare, color="b")
```

## Diagramme camenbert (pie)

```python
plt.pie(titanic.Fare, autopct='%1.1f%%', startangle=180)
```

## Nuage de points

```python
plt.scatter(titanic.PassengerId, titanic.Fare)
```

## Empiler plusieurs graphes différents

```python
plt.scatter(titanic.PassengerId, titanic.Fare)
plt.bar(titanic.PassengerId, titanic.Fare, color="y")
plt.plot(titanic.PassengerId, titanic.Age, "c", linewidth=10)
```
