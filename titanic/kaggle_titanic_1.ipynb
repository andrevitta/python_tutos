## Jeu de test

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler

train = pd.read_csv("./data/train.csv")
test = pd.read_csv("./data/test.csv")
```

## Création d'une fonction de préparation (à partir des éléments précédents)

```python
def dataprep(data):
    sexe = pd.get_dummies(data['Sex'], prefix='sex')
    cabin = pd.get_dummies(data['Cabin'].fillna('X').str[0], prefix='Cabin')
    age = data['Age'].fillna(data['Age'].mean())
    emb = pd.get_dummies(data['Embarked'], prefix='emb')
    
    scaler = MinMaxScaler()
    fare = data['Fare'].fillna(0) # Une donnée de test n'a pas de Prix !
    faresc = pd.DataFrame(scaler.fit_transform(data[['Fare']].fillna(0)))

    dp = data[['Pclass', 'SibSp']].join(sexe).join(emb).join(faresc).join(cabin).join(age)
    return dp
```

```python
Xtrain = dataprep(train)
# remove the Cabin with T value as this does not exist in the test dataset !
del Xtrain['Cabin_T']
Xtest = dataprep(test)
```

```python
Xtrain.head(3)
```

## Entrainement du modèle

```python
y = train.Survived
rf = RandomForestClassifier(n_estimators=100, random_state=0, max_features=2)
rf.fit(Xtrain, y)
p_tr = rf.predict(Xtrain)
print ("Score Train -->", round(rf.score(Xtrain, y) *100,2), " %")
```
